```python
#reproj_env
import os
import rioxarray as rio
import xarray as xr
```

#### Reproject NetCDF files to UTM grid

```python
sm_data = rio.open_rasterio('W:/VUB/_main_research/mHM/mhm_outputs/mhm_RMI/mhm_mon_fluxes.nc')
sm_xarray = xr.open_dataset('W:/VUB/_main_research/mHM/mhm_outputs/mhm_RMI/mhm_mon_fluxes.nc')
```

```python
# Assign the original CRS
UTM_projection = "EPSG:32631"  # UTM Zone 31N
sm_xarray.rio.set_spatial_dims(x_dim="lon", y_dim="lat", inplace=True)
sm_xarray.rio.write_crs("EPSG:4326", inplace=True)

# Step 2: Reproject to EPSG:32631 (UTM Zone 31N)
ds_proj = sm_xarray.rio.reproject(UTM_projection, resolution=2000)

ds_proj = ds_proj.rename({"x": "lon", "y": "lat"})

#change the longname of the coordinates
ds_proj.lon.attrs
```
